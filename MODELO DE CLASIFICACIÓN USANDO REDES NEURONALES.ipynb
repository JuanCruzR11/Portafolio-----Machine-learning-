```python
X = df.drop("benign_0__mal_1",axis=1).values
y = df["benign_0__mal_1"].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
#hasta aca hicimos análisis exploratorio de datos, seperamos los datos y los normalizamos
```

## PARTE 2 

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense,Dropout
```

```python
X_train.shape #Tamaño de los datos
```

```python
model = Sequential()

model.add(Dense(30,activation="relu"))

model.add(Dense(15,activation="relu"))


#BINARY CLASSIFICACTION
model.add(Dense(1,activation="sigmoid"))

model.compile(loss="binary_crossentropy",optimizer="adam")
```

```python
model.fit(x=X_train,y=y_train,epochs=600,validation_data=(X_test,y_test))
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

```python
#CUANDO LA LINEA NARANJA INCREMENTA MIENTRRAS LA AZUL AUN ESTA BAJANDO QUIERE DECIR QUE ESTAMOS AJUSTANDO DEMASIADO LOS DATOS DE ENTRENAMIENTO, ES DECIR, MUCHAS EPOCHS
```

```python
model = Sequential()

model.add(Dense(30,activation="relu"))

model.add(Dense(15,activation="relu"))


#BINARY CLASSIFICACTION
model.add(Dense(1,activation="sigmoid"))

model.compile(loss="binary_crossentropy",optimizer="adam")
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(monitor="val_loss",mode="min",verbose=1,patience=25) 
#Le ponemos el modo minimizar porque estamos hablando de la pérdida
```

```python
model.fit(x=X_train,y=y_train,epochs=600,validation_data=(X_test,y_test),
         callbacks=[early_stop])
```

```python
model_loss = pd.DataFrame(model.history.history)
```

```python
model_loss.plot() #AHORA ESTAN MUCHO MAS CERCANAS
```

```python
from tensorflow.keras.layers import Dropout
```

```python
model = Sequential()

model.add(Dense(30,activation="relu"))
model.add(Dropout(0.5))

model.add(Dense(15,activation="relu"))
model.add(Dropout(0.5))


#BINARY CLASSIFICACTION
model.add(Dense(1,activation="sigmoid"))

model.compile(loss="binary_crossentropy",optimizer="adam")
```

```python
model.fit(x=X_train,y=y_train,epochs=600,validation_data=(X_test,y_test),
         callbacks=[early_stop])
```

```python
model_loss = pd.DataFrame(model.history.history)
```

```python
model_loss.plot() #MEJORÓ AUN MAS USANDO EL APAGADO DE NEURONAS
```

```python
y_pred = model.predict(X_test) 
```

```python
y_class = np.where(y_pred > 0.5, 1,0)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,y_class))
print("\n")
print(confusion_matrix(y_test,y_class))
```
